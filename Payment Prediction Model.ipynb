# Preprocessing

### Date Time Conversion

```python
# Converting into datetime datatype and assign all the conversion into format ='YYYY-MM-DD'.

main_train['document_create_date'] = pd.to_datetime(main_train['document_create_date'],format='%Y%m%d')
main_train['document_create_date.1'] = pd.to_datetime(main_train['document_create_date.1'],format='%Y%m%d')
main_train['clear_date'] = pd.to_datetime(main_train['clear_date'])
main_train['clear_date'] = pd.to_datetime(main_train['clear_date'],format='%Y%m%d')
main_train['posting_date'] = pd.to_datetime(main_train['posting_date'])
main_train['posting_date'] = pd.to_datetime(main_train['posting_date'],format='%Y%m%d')
main_train['due_in_date'] = main_train['due_in_date'].astype(int)
main_train['baseline_create_date'] = main_train['baseline_create_date'].astype(int)
main_train['due_in_date'] = pd.to_datetime(main_train['due_in_date'],format='%Y%m%d')
main_train['baseline_create_date'] = pd.to_datetime(main_train['baseline_create_date'],format='%Y%m%d')
main_train.head()
```

```python
main_train.info()
```

#### Delay is the target column

```python
# delay = cleardate - due_in_date (target)
main_train['Delay'] = main_train['clear_date'] - main_train['due_in_date']
main_train['Delay'] = main_train['Delay'].dt.days
main_train
```

```python
# Since there are only two types of currency:- USD and CAD so converting the total_open_amount in USD only.
temp = main_train[['total_open_amount', 'invoice_currency']].copy()
temp.head()
```

```python
# Funtion to convert currency.
def convert_to_usd(args):  
    amount, currency = args
    rates = {'USD': 1.23, 'CAD': 0.81}  # Convertion [1 USD = 1.23 CAD or 1 CAD = 0.81 USD]
    return rates[currency] * amount

temp['In_USD'] = temp.apply(convert_to_usd, axis=1)
```

```python
# Updating the total_open_amount column.
main_train['total_open_amount'] = temp['In_USD'].copy()
```
